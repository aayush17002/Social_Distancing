```python
import cv2
import numpy as np
import time
import math 
# Create our body classifier
body_classifier = cv2.CascadeClassifier('haarcascade_fullbody.xml')

# Initiate video capture for video file
cap = cv2.VideoCapture('walking.avi')

# Loop once video is successfully loaded
frame_no=0

while cap.isOpened():
    print("Frame number = ",frame_no)
    frame_no+=1
    
    # Read first frame
    ret, frame = cap.read()
    #frame = cv2.resize(frame, None,fx=0.5, fy=0.5, interpolation = cv2.INTER_LINEAR)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Pass frame to our body classifier
    bodies = body_classifier.detectMultiScale(gray, 1.2, 3)
    
    # Extract bounding boxes for any bodies identified
    idx=0
    
    L=[]
    for (x,y,w,h) in bodies:
#         frame = cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 255), 2)
        string="Box {0}: ({1},{2}), ({3},{4}), ({5},{6}), ({7},{8})".format(idx,x,y,x+w,y,x+w,y+h,x,y+h)
#         new approach
        x_mid = (x+x+w)/2
        y_mid = (y+y+h)/2
        z = ((165*658)/abs(math.sqrt(w*w + h*h)))
        x_mid2 = math.sqrt(abs(z*z - x_mid*x_mid))
        y_mid2 = math.sqrt(abs(z*z - y_mid*y_mid))
        L.append([x_mid2,y_mid2,z,x,y,h,w])
        
#         L.append([(x+x+w)//2,(y+y+h)//2,((165*658)/abs(y-(y+h))),x,y,h,w])
#         cv2.putText(frame, string, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (36,255,12), 2)
        cv2.imshow('Pedestrians', frame)
        idx+=1
    
    for i in range(len(L)-1):
        for j in range(i+1,len(L)):
            x_net=L[i][0]-L[j][0]
            y_net=L[i][1]-L[j][1]
            val = ((165*658)/abs(math.sqrt(x_net*x_net + y_net*y_net)))
            
#             z_net=L[i][2]-L[j][2]
#             val = math.sqrt((x_net*x_net) + (y_net*y_net) + (z_net*z_net))
            
#             if(val*0.0328<15):
            if(val*0.0328<6):
                frame_out_i = cv2.rectangle(frame, (L[i][-4], L[i][-3]), (L[i][-4]+L[i][-1], L[i][-3]+L[i][-2]), (0, 0, 255), 1)
                frame_out_j = cv2.rectangle(frame, (L[j][-4], L[j][-3]), (L[j][-4]+L[j][-1], L[j][-3]+L[j][-2]), (0, 0, 255), 1)

                cv2.imshow('Pedestrians',frame_out_i)
                cv2.imshow('Pedestrians',frame_out_j)

#                 s = "distance between " + str(i) + " and " + str(j) + " : " + str(round(val*0.0328,3))+ "  x_net= " +str(round(x_net*0.0328,3))+ "   y_net= "+str(round(y_net*0.0328,3))+ "    z1= "+str(round(L[i][2]*0.0328,3))+"   z2= "+str(round(L[j][2]*0.0328,3))+ "    z_net= "+str(round(z_net*0.0328,3))    
                s = "distance between " + str(i) + " and " + str(j) + " : " + str(round(val*0.0328,3))
                print(s)

#     key=cv2.waitKey(500)
    
    if cv2.waitKey(1) == 13: #13 is the Enter Key
        break
    
cap.release()
cv2.destroyAllWindows()
```
